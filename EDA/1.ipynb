GRAPH INSIGHTS

Most tip values are small, and only a few are high.

Both color groups follow almost the same pattern.

One color appears slightly more frequent, but the difference is not big.

The tips for both groups are mostly concentrated in the same range.

Very high tip values are rare for both colors.


**Hypotheses:**  
- H0: distribution of tip for yellow = distribution of tip for green  
- H1: distribution of tip for yellow â‰  distribution of tip for green

```python
yellow_tip=df['tip'][df['color']=='yellow']
green_tip=df['tip'][df['color']=='green']
```

```python
yellow_tip.shape
```

```python
green_tip.shape
```

```python
# Independent T-Test

from scipy import stats
t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
#print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
alpha=0.05 #significance label
if p_val < alpha:
    print("Reject null hypothesis:The tip for yellow category is different from that of green")
else:
    print("Fail to reject null hypothesis")
```

```python
cash_payment=df['tip'][df['payment']=='cash']
card_payment=df['tip'][df['payment']=='credit card']
```


**Hypotheses:**  
- H0: distribution of tip through cash = distribution of tip through credit card  
- H1: distribution of tip  through cash â‰  distribution of tip through credit card  

```python
# Independent T-Test

from scipy import stats
t_stat, p_val = stats.ttest_ind(cash_payment, card_payment)
#print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
alpha=0.05 #significance label
if p_val < alpha:
    print("Reject null hypothesis:The tip for payment through card is different from that of cash")
else:
    print("Fail to reject null hypothesis")
```

**Hypothesis**

```python
manhatten_tip=df['tip'][df['pickup_borough']=='Manhattan']
bronx_tip=df['tip'][df['pickup_borough']=='Bronx']
brooklyn_tip=df['tip'][df['pickup_borough']=='Brooklyn']
queens_tip=df['tip'][df['pickup_borough']=='Queens']
staten_tip=df['tip'][df['pickup_borough']=='Staten ']
```

```python
#ANOVA test

from scipy import stats
t_stat, p_val = stats.f_oneway(manhatten_tip,bronx_tip,brooklyn_tip,queens_tip)

print("P-value:", p_val)
```

```python
alpha=0.05 #significance label
if p_val < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

**HYPOTHESIS**

H0: correlation is zero
H1 : there is significant correlation (not zero)

```python
r,p_val=stats.spearmanr(df['tip'],df['passengers'])
print("correlation :",r)
print("p-value:",p_val)
alpha=0.05 #significance label
if p_val < alpha:
    print("Reject null hypothesis.There is no relationship")
else:
    print("Fail to reject null hypothesis.There is a relationship")
```
